生成一个矩阵，通过matplotlib绘制出来散点图

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
tf.compat.v1.disable_eager_execution()
a = tf.random.normal([2,30])
sess = tf.compat.v1.Session()
out = sess.run(a)
x,y = out
plt.scatter(x, y)
plt.show()
```
